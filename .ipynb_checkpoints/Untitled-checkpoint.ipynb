```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import time
df = pd.read_csv("main.csv", parse_dates = ["date"], index_col = 0)
```

```python
df = pd.read_csv("cumsum.csv", parse_dates = ["date"], index_col = 0)

y = 1
aveReturn = []
aveVolatility = pd.DataFrame(columns = ['Size', 'Volatility'])

while y < 1000: 
    randomcryptos = random.randint(2,10)
    cryptos = random.sample(range(1,880), randomcryptos)
    randomday = random.sample(range(1, 333), 1)
    portfolioSize = len(cryptos)
    #weights = random.random(10)
    data = df.pct_change()
    return_data = data.iloc[:,cryptos].dropna(axis = 1, how = 'all')
    returns = return_data.iloc[randomday].sum()
    volatility = return_data.iloc[randomday].std()
    #std_data = returns.std()
    #volatility = std_data.iloc[randomday].sum()

    aveReturn.append(returns)
    aveVolatility = aveVolatility.append({'Size': portfolioSize, 'Volatility': volatility}, ignore_index = True)
    y +=1
```

```python
df = aveVolatility
df
```

```python
df = df.sort_values(by = 'Size')
df.set_index('Size', inplace = True)
df.head()
```
